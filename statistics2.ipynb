### mixed models (lmer(y~x+(1|x)))
important for datasets with multiple subjects and multiple items

```r
library(languageR)
data(lexdec)
head(lexdec)
#an experiment in which every subject responded to every word
```

```r
lexdec$RT = -1000/exp(lexdec$RT) #inverse transform the log-transformed RT
```

to treate item and subject as additional sources of random noise  
**fixed-effect factors**: small number of levels, levels are experimentally replicable  
**random-effect factors**: sources of noise tied to observational units, observational units are randomly selected, hence experimentally not replicable

```r
#fit a linear mixed model
library(lme4)
lexdec.lmer = lmer(RT ~ SubjFreq + NativeLanguage +
(1|Subject) + (1|Word), data=lexdec)
summary(lexdec.lmer)
```

##### model coefficients
intercept = −1.508 : the estimate for an unknown subject  
slope = −0.084 for SubjFreq  
slope = 0.243 for NativeLanguage (contrast between "English" and "Other")  
the stdev of the residual error = 0.269  
the stdev of the subject random noise on the intercept = 0.219  
the stdev of the word random noise on the intercept = 0.082  
The LMM uses the estimates of standard deviations (or variances) of the random-effect factors to derive estimates of by-subject or by-item adjustments. These adjustments can be extracted by ranef(lexdec.lmer)$Word  
shrinkage protects us against overfitting

```r
#more complex random effect structure
lexdec.lmer2 = lmer(RT ~ SubjFreq + NativeLanguage +
(1+SubjFreq|Subject) + (1|Word), data=lexdec)
summary(lexdec.lmer2)
#fast subjects (upper left) show little of a frequency effect
#whereas slow subjects show the largest frequency effect
```

```r
#use anova to compare models
anova(lexdec.lmer, lexdec.lmer2, refit=FALSE)
```

for data where the response variables are counts, or proportions

###### types of count data
**logistic regression**: binary response variables, counts of successes and failures (e.g. counts of correct vs. incorrect responses in an experiment; counts of NP vs. PP constructions)  
**Poisson regression**: counts of observations, counts of how often an event occurs in a fixed interval of 'time' (e.g. number of particles emitted in a fixed time interval; the number of occurrences of "the" in a text with N tokens)

if we have data in the form of proportions or percentages, we have heteroskedasticity, and using the linear model becomes risky  
furthermore, probabilities are bounded between 0 and 1, and percentages are bounded between 0 and 100, but the standard regression set-up presupposes the response ranges from minus infinity to plus infinity

**logits**(log odds ratios)=log(successes/failures)  
positive when there are more successes than failures  
0 when there are as many successes as failures

##### Logistic Regression

```r
library(languageR)
data(dative)
table(dative$RealizationOfRecipient)
head(dative)
```

```r
dative.glm = glm(RealizationOfRecipient ~
AnimacyOfRec + AnimacyOfTheme +
DefinOfRec + DefinOfTheme +
PronomOfRec + PronomOfTheme +
AccessOfRec + AccessOfTheme +
LengthOfRecipient + LengthOfTheme,
data=dative, family="binomial")
summary(dative.glm)
```

method of maximum likelihood: maximum likelihood estimate of p in binomial model is y/n , which is the proportion of successes in n trials  
**deviance residuals**: should be roughly normal, outliers are to be carefully inspected  
**null deviance**: the deviance of an intercept-only model(if the model is defined as y~1, then null deviance and residual deviance would be the same)  
**residual deviance**: the deviance of the model fitted to the data, with respect to the saturate model (a perfectly-fitted model), the smaller, the better the fit     
**AIC**:  Deviance + 2 × the number of parameters  
**Fisher Scoring iterations**: number of iterations required to obtain parameter estimates that are precise enough

```r
#test whether the model fits the data
1-pchisq(deviance(dfr.glm), df.residual(dfr.glm))
```

```r
#compare two models using anova
anova(dfr.glm2, dfr.glm, test="Chisq")
```

**lm**: linear regression model, 1 regression line (or plane/hyperplane)  
**lmer**: repeated measures, if we have k participants responding to w words, k × w observations:  
k regression lines, one for each subject  
if we just fit 1 regression line separately to the data of each subject, we overfit -
solution: shrinkage  
subjects can vary with respect to the intercept  
and can also vary with respect to the slope of the regression line  
**glm**: generalized linear model:  
binomial (proportions/percentages)  
poisson (counts)  
**glmm**: generalized linear mixed model  
**AIC**: balances the complexity of the model (represented by its number of parameters) against its deviance

#### GLM: Poisson regression (glm(y~x, family="poisson"))
count data are positive, and the variance varies with the mean, so a linear model of the count itself is inappropriate (might produce negative predicted counts)
we therefore model the log of the rate parameter as a linear function of the predictors: 
$log(\lmbda) = \beta0 + x_i\beta$

```r
#predict accuracy(number of correct responses made out of a total of 30) 
#from written frequency and age of subject
library(languageR)
data(english)
head(english)
```

```r
english.glm = glm(CorrectLexdec ~ WrittenFrequency * AgeSubject,
data = english,
family = "poisson")
summary(english.glm)
```

- for old subjects, the slope is 0.025 (so accuracy increases with frequency)
- for young subjects, the intercept is adjusted downwards by 0.22 (they are less accurate at log freq = 0), the slope of frequency is adjusted upward by 0.028, so it more than doubles compared to the old subjects
  
  there is no reason to suppose that the model is inappropriate for the data

### gam (Generalized Additive Models)
polynomials: fit wiggly curves with polynomials; increase the goodness of the fit by increasing the degree  
however: the fit can look good, but it goes wild outside the interval where there is data; the fit may also go wild for isolated individual data points  

**GAMs** require fewer parameters and don't have this problem
$$y_i = \beta_0 + f_q(x_i) + noise$$
the regression smooth fq(x) is a weighted sum of a set of q basis functions defined over the predictor x

```r
library(languageR)
library(mgcv)
data(english)
english.gam = gam(RTlexdec ~ s(WrittenFrequency),
data=english[english$AgeSubject=="young", ])
plot(english.gam, xlab="log word frequency", ylab="partial effect")
```

the plot shows the partial effect of frequency, the curve is calibrated to be centered around the x-axis

```r
summary(english.gam)
```

the intercept has to be added to get the predicted RTs

avoid overfitting by penalizing complexity (wiggliness)
- for each basis function, we calculate the shrinkage factor (e.g., a shrinkage factor 0.5 indicates that the original weight has been reduced by half)
- the sum of the shrinkage factors is referred to as the effective degrees of freedom (edf)
- if the edf is close to 1, then most wiggliness has been penalized away
- importantly, if a predictor has a linear effect, a GAM will find out that and will not report a non-linear effect

1 edf indicates linearity  
smoothing splines: weighted sums of basis functions  
by-smooths (one smooth for every factor level)

k specifies the maximum number of basis functions that the model
should consider

```r
#a baseline model without interaction
pupil.gam0 = gam(Dilation ~ Condition + s(Time), data = pupil)
#add the interaction with the ‘by’-smooth
pupil.gam = gam(Dilation ~ Condition + s(Time, by = Condition),
data = pupil)
```

```r
#interactions of factor and two numerical predictors
#a baseline model without the three-way interaction
eggbox.gam0 = bam(Z ~ Condition + te(X, Y), data=eggbox)
eggbox.gam1 = bam(Z ~ Condition + te(X, Y, by = Condition), data = eggbox)
```

##### GAMMs (mixed GAMs)
random effects are modeled with "re" smooths

```r
library(languageR)
data(lexdec)
# instead of log-transformed RTs, we use inverse-transformed RTs
lexdec$RTinv = -1000/exp(lexdec$RT)
lexdec$Frequency = as.numeric(scale(lexdec$Frequency))
lexdec$Trial = as.numeric(scale(lexdec$Trial))
head(lexdec)
```

```r
library(mgcv)
lexdec.gamm = gam(RTinv ~ Frequency + Trial +
s(Subject, bs="re") +
s(Word, bs="re"),
data = lexdec)
summary(lexdec.gamm)
```

```r
lexdec.gamm3 = gam(RTinv ~ s(Frequency) + s(Trial) +
s(Subject, bs="re") +
s(Word, bs="re"),
data = lexdec)
summary(lexdec.gamm3)
```

```r
c(AIC(lexdec.gamm), AIC(lexdec.gamm3))
```

```r
#a factor smooth for trial
lexdec.gamm4 = gam(RTinv ~ s(Frequency) +
s(Trial, Subject, bs='fs', m=1) +
s(Word, bs="re"),
data = lexdec)
summary(lexdec.gamm4)
```

```r
c(AIC(lexdec.gamm3), AIC(lexdec.gamm4))
#this made the model a lot more precise
```

```r
options(repr.plot.width=5, repr.plot.height=5)
plot(lexdec.gamm4, select=2)
#select=2 means to select the second row of smooth terms, which are s(Trial, Subject)
```
